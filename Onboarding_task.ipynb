#### 1. Представьте, что вы работаете над онбордингом пользователей в https://app.wazzup24.com. Какие продуктовые метрики выберете для этой задачи? Почему именно их? Как будете их считать? Была бы какая-то из них для вас главной и почему? 

1. Time to value
- Это время, которое пользователь должен потратить до момента получения ценности от продукта. Чем меньше это значение тем лучше наш Onboarding  
важно определить какой ивент будет считаться завершением Onboarding: в нашем случае это может быть достижение Шага: Choose your pricing plan
- эта метрика дает представление о том, как много усилий пользователю необходимо пройти до этапа, когда пользователь понимает ценность продукта
- для расчета необходжимо фиксировать дату-время лида и дату время всех последующих этапов Onboarding,   
$$TimeToValue = time_{ValueAction} - time_{Onboarding}$$

2. Product adoption success rate
- Главная метрика в случае анализа Онбординга, так как дает представление о том, насколько онбординг справляется со своей главной задачей - Активацией пользователей
- Product adoption - это результат процесса Onboarding.
- Чтобы зафиксировать этот результат нам нужно определить этап Онбординга, который будет говорить нам: "Супер, у нас есть новый пользователь!"

- В нашем случае мы можем ввести пробный период использования и измерять free-to-paid conversion (конверсию от бесплатной версии к оплате подписки)
- Такая метрика позволит определять как изменения в Онбординге и Маркетинге влияют на эффективность


3. Retention rate
- Продукт будет более успешным если сможет не только привлекать новых пользователей, но и удерживать текущих.
- Мы можем измерять Ratention Rate наблюдая за количеством пользователей в нашем сервисе в течение определенного времени (день, месяц, год)
- Мы также можем измерять этот показатель среди Кохорт (по времени регистрации, девайсам, локации и т.д.) (Кохортный анализ) это позволит увидеть реакцию пользователей на введение изменений в процесс онбординга

4. Funnel Analysis
- Мы можем ввести Воронку, чтобы находить на каком этапе пользователи "отваливаются" из процесса онбординга.
- Мы можем измерять конверсию на каждой кнопке или форме, которая должна быть заполнена
- Воронки можно также строить по Кохортам, чтобы определить какие сложности возникают у разных групп пользователей

4. Churn Rate
- Отток показывает как часто пользователи покидают продукт.
- Эта метрика важна, так как одна из главных задач Онбординга помочь пользователю извлечь ценность из продукта для себя, благодаря чему у пользователя появится желание вернуться
$$ \frac{Кол-во. польз. ушедших}{Кол-во. общее. польз. за. период} $$

5. User support request rate
- Если пользователь отправляет запрос на поддержку, значит что-то пошло не так.
- Задача снизить к минимуму количество пользователей, которые выпадают из тропы по которой их ведут дизайнеры продукта
- Следовательно мы можем оценить успех процесса онбординга просто посчитав количество Тикетов на поддержку или FAQ

1. Онбординг не нужен в принципе :)  
- Мы смотрим на воронку этапов в Онбординге и видим снижение на каждом этапе. 
- Возникает гипотеза, может быть процесс онбординга в целом снижает конверсию на оплату - больше стадий больше барьеров, пусть будет зарегистрировался - оплатил - используешь  
- Такая гипотеза имеет самый высокий приоритет, так как может быть ситуация когда мы решаем проблему, которую сами и создали

2. Давайте подробно расскажем про продукт:
- Скриншоты
- Видео
- Анимации (как у нас все внутри классно работает)
- Похвастаемся как мы сотрудничаем с крупными компаниями
Приоритет 2 - это классический подход, который сразу приходит на ум Продактам при разработке Онбординга

3. Пойдем в психологию:
- А что если задать пользователю вопрос?
- Что если это работает лучше чем картинка?
    - Что если сделать линк между ответами пользователей и следующим вопросом (подводка). Другими словами: "хмм раз ты отвечаешь так, что тогда будет вот так"
Приоритет 3 - Когда мы задаем пользователю вопрос, связанный с продуктом, у него в голове может выстраиваться картинка - раз меня сейчас это спрашивают, значит потом это где-то будет нужно  
Но, здесь есть обратная сторона, если это правило не соблюдается (мы спрашиваем что-то чего в продукте нет - у нас может резко вырасти Churn подписок)  
У пользователя появляется понимание чем ему будет полезен продукт  
- PayWall - Оплата - С Trial или нет, но разница с гипотезой 2 - У тебя как у пользователя сформировались собственные ожидания от того что будет в продукте, а не Product тебе дает их на блюдечке готовые    
- У пользователя возникает ощущение что это уже именно про него

4. Что если ответы на вопросы пользователя будут приводить его в разные ветки Онбординга, это позволит выработать различные сценарии онбординга и лучше адаптировать его по определенные сегменты

5. Такие Generic (крупные эксперименты) можно разбить на большое колиечство практически атомарных экспериментов:
- Мы даем один вопрос без сценариев
- Один вопрос и два сценария в зависимости от ответа
- Добавляем вопросы один за одним и следим за конверсией
- Меняем местами вопросы и следим за конверсией

#### Продуктовая команда выяснила, что в последние месяцы конверсия из тестирования продукта на бесплатном пробном периоде в покупку подписки сильно упала. Вам нужно выяснить возможные причины. Как вы подойдете к исследованию этой проблемы?

- Что в целом происходит с рынком?
    - Возмонжно проблема лежит за пределами сервиса (экономика, конкуренты, блокировка платежных систем и т.д.)
    - Помочь понять может динамика конкурентов (Yandex Google)
- Какие были проведены изменения в последнее время?
    - В процессе оплаты?
    - В функционале бесплатной версии?
    - На этапе регистрации? - возможно часть данных перенесли из регистрации на стадию оплаты
    - Pricing (новые условия - изменение цены или убрали старый вариант подписки)  
- Что пользователи отвечают на вопрос почему они не перешли к оплате? - Главное понять истинную причину

Нужна сквозная аналитика с данными по трафику, продажам и кликам на всех этапах
